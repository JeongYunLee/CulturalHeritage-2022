```python
df_1 = df_tmp[['문화재명_지정호수', '문화재분류1', '문화재분류2', '문화재분류3', '문화재분류4', '재질', '재질상세', '시대_1', '시대구분_1', '시대구분_2', '소재지시도_시명', '소재지도명', '출토지시도_시명', '출토지도명']]
df_1
```

```python
## fillna
df_1.fillna('NaN', inplace=True)

id_list = []
column_list = list(df_1.columns)
for i in tqdm(range(len(df_1))):
  for j in range(len(column_list)):
    id_list.append(df_1[column_list[j]][i])
  
num_list=[]
for i in range(0,351):
  for j in range(0,14):
    num_list.append(i)

# group_list=[]
# for i in range(1,352):
#   for j in range(0,14):
#     group_list.append(j)
  
col_list = []
col_list.append(column_list * 351)
col_list = sum(col_list, [])
```

```python
## dataframe
df_nodes = pd.DataFrame({'id':id_list, 'group':'null', 'hover':col_list})
df_nodes = df_nodes[df_nodes['id']!='NaN']
df_nodes.drop_duplicates(subset=['id'], inplace=True)
df_nodes.reset_index(inplace=True, drop=True)

## group 번호로 바꿔주기 
df_nodes['group'].loc[df_nodes['hover']=='문화재명_지정호수'] = '1'
df_nodes['group'].loc[df_nodes['hover'].str.contains('문화재분류')] = '2'
df_nodes['group'].loc[df_nodes['hover'].str.contains('재질')] = '3'
df_nodes['group'].loc[df_nodes['hover'].str.contains('시대')] = '4'
df_nodes['group'].loc[df_nodes['hover'].str.contains('소재지')] = '5'
df_nodes['group'].loc[df_nodes['hover'].str.contains('출토지')] = '6'

## json
json_nodes = df_nodes.to_json(orient = 'records',force_ascii=False)

df_nodes
```

### links (df to json)

```python
## dataframe
df_nodes_tmp = pd.DataFrame({'id':id_list, 'num':num_list, 'column':col_list})
df_nodes_tmp = df_nodes_tmp[df_nodes_tmp['id']!='NaN']
df_nodes_tmp.reset_index(inplace=True, drop=True)

df_nodes_tmp
```

```python
## 문화재명(국문) 행만 삭제하기 
df_links_tmp = df_nodes_tmp[df_nodes_tmp['column']!='문화재명_지정호수']

## source 컬럼 추가하기 
df_source = df_tmp[['문화재명_지정호수', 'Column']]
## group 번호를 기준으로 merge
df_links = pd.merge(df_source, df_links_tmp, left_on='Column', right_on='num')

## 컬럼 이름 바꾸기
df_links = df_links[['문화재명_지정호수', 'id', 'column']]
df_links.columns=['source', 'target', 'value']

########### 계층 구조 만들어주기 ##############
### 문화재분류 (4->3->2->1)
df_links_class_tmp = df_links[df_links['value'].str.contains('문화재분류')]
df_links_class_tmp.reset_index(inplace=True, drop=True)
df_links_class = df_links_class_tmp.copy()

for i in tqdm(range(len(df_links_class_tmp))):
  try:
    if df_links_class_tmp['value'][i] == '문화재분류4':
      pass
    elif df_links_class_tmp['value'][i] == '문화재분류3':
      df_links_class['source'][i] = df_links_class_tmp[(df_links_class_tmp['source']==df_links_class_tmp['source'][i])&(df_links_class_tmp['value']=='문화재분류4')]['target'][i+1]
    elif df_links_class_tmp['value'][i] == '문화재분류2':
      df_links_class['source'][i] = df_links_class_tmp[(df_links_class_tmp['source']==df_links_class_tmp['source'][i])&(df_links_class_tmp['value']=='문화재분류3')]['target'][i+1]
    elif df_links_class_tmp['value'][i] == '문화재분류1':
      df_links_class['source'][i] = df_links_class_tmp[(df_links_class_tmp['source']==df_links_class_tmp['source'][i])&(df_links_class_tmp['value']=='문화재분류2')]['target'][i+1]
    else:
      pass
  except:
    pass

### 재질 (재질상세->재질)
df_links_material_tmp = df_links[df_links['value'].str.contains('재질')]
df_links_material_tmp.reset_index(inplace=True, drop=True)
df_links_material = df_links_material_tmp.copy()

for i in tqdm(range(len(df_links_material_tmp))):
  try:
    if df_links_material_tmp['value'][i] == '재질상세':
      pass
    elif df_links_material_tmp['value'][i] == '재질':
      df_links_material['source'][i] = df_links_material_tmp[(df_links_material_tmp['source']==df_links_material_tmp['source'][i])&(df_links_material_tmp['value']=='재질상세')]['target'][i+1]
    else:
      pass
  except:
    pass

### 나머지
df_links_others = df_links[~df_links['value'].str.contains('문화재분류')]
df_links_others = df_links_others[~df_links_others['value'].str.contains('재질')]

## 최종 정리 
df_links = pd.concat([df_links_others, df_links_material, df_links_class], axis=0)
df_links.drop_duplicates(inplace=True)
df_links.reset_index(inplace=True, drop=True)
df_links
```
