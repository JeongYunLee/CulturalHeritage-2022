## ⭕ 소재지 
- 소재지가 두개인 경우 1개 있음
- 소재지 상세에서 시군구 명이 '서울' 이렇게 적혀 있는 것들 모두 '서울특별시'로 바꿔줌
- 기본규칙 `시도 + 시군구 + 읍면 + 도로명 + 건물번호(본번 - 부번) + (동, 건물명)`
  - 읍면 인 경우 : 시도 + 시군구 + 읍면 + 도로명 + 건물번호 + (건물명)
	→ 시군구 뒤의 ‘읍/면’ 으로 끝나는 단어가 읍면 이름입니다.    
	→ 예: 경상북도 안동시 서후면 봉정사길 222, 봉정사 (태장리) → 서후면    
  - 동 인 경우 : 시도 + 시군구 + 도로명 + 건물번호 ~ (동, 건물명) 
	→ 맨 뒤의 괄호 안의 단어가 동 이름입니다.     
	→ 예: 서울 중구 세종대로 40 (남대문로4가) → 동 이름 : 남대문로4가

```python
## 개요 통일시키기
for i in range(len(df)):
  if '개요' not in df['설명 텍스트'][i]:
    print(i)
    print(df['지정호수'][i])
    print(df['설명 텍스트'][i])
```

```python
## 제작연대 통일시키기
for i in range(len(df)):
  df['설명 텍스트'][i] = df['설명 텍스트'][i].replace('제작 연대', ' 제작연대').replace('제작연도', ' 제작연대').replace('제작 시기', ' 제작연대').replace('제작연대(제작지)', ' 제작연대').replace('제작연대(제작자)', ' 제작연대')

  if '제작연대' not in df['설명 텍스트'][i]:
    print(df['지정호수'][i])
    print(df['설명 텍스트'][i])
```

```python
## 특징 - 크기 및 형태 통일시키기
for i in range(len(df)):
  df['설명 텍스트'][i] = df['설명 텍스트'][i].replace('–', '-').replace('특징 -크기 및 형태', '특징 - 크기 및 형태').replace('특징 - 크기 및 형태, 용도', '특징 - 크기 및 형태')

  if '특징 - 크기 및 형태' not in df['설명 텍스트'][i]:
    ## 크기 및 형태 --> 이렇게만 써준 경우 '특징 - 크기 및 형태' 로 바꾸기
    df['설명 텍스트'][i] = df['설명 텍스트'][i].replace('크기 및 형태', '특징 - 크기 및 형태')
    if '특징 - 크기 및 형태' not in df['설명 텍스트'][i]:
      print(df['지정호수'][i])
      print(df['설명 텍스트'][i])
```

```python
## 특징 - 내용 및 의의 통일시키기
for i in range(len(df)):
  df['설명 텍스트'][i] = df['설명 텍스트'][i].replace('특징 및 의의', '내용 및 의의').replace('특징 - 의의와 평가', '특징 - 내용 및 의의').replace('특징 - 화법 및 평가', '특징 - 내용 및 의의').replace('내용 및 평가', '내용 및 의의')

  if '특징 - 내용 및 의의' not in df['설명 텍스트'][i]:
    ## 내용 및 의의 --> 이렇게만 써준 경우 '특징 - 내용 및 의의' 로 바꾸기
    df['설명 텍스트'][i] = df['설명 텍스트'][i].replace('내용 및 의의', '특징 - 내용 및 의의')
    if '특징 - 내용 및 의의' not in df['설명 텍스트'][i]:
      print(df['지정호수'][i])
      print(df['설명 텍스트'][i])
```

```python
## 기타 통일시키기

for i in range(len(df)):
  if '기타' not in df['설명 텍스트'][i]:
    ## 기타 카테고리 없는 경우 문단 가장 마지막에 '기타' 추가해주기
    df['설명 텍스트'][i] = df['설명 텍스트'][i] + '\n \n 기타 \n'
    if '기타' not in df['설명 텍스트'][i]:
      print(df['지정호수'][i])
```
